
RGB图像

色彩通常由三个独立的颜色通道组成：红色（Red）、绿色（Green）和蓝色（Blue）。

每个通道的取值范围通常是0到255（或0到1），其中0表示完全无颜色，255表示完全饱和的颜色。

每个像素的颜色由这三个基本颜色组合而成，因此，一个图像在二维平面上看似只有一个像素，实际上是由三个不同颜色（不同通道）的像素叠加而成的。

![一张彩色RGB图片按照通道维度C堆叠](./learn-images/RGB.webp)

通过调整红，绿，蓝三个通道的数值，我们可以改变图像的颜色。

每个通道的颜色值范围是0-255，分别代表红，绿，蓝的强度。

彩色图片有RGB 三个通道，如果将三张分别为红色、绿色、蓝色的图片叠加在一起，就可以得到彩色图片。

同样的，也可以将彩色图片的三个通道分别提取出来，得到三张分别为红色、绿色、蓝色的图片。

RGB 图像在计算机中的占用大小

    - 彩色图像的占用大小 = 图像的宽度 * 图像的高度 * 3   （每个通道的每个像素占用 1 个字节）
    
    - 如一个3通道的 rgb图像，长和宽分别为 1920x1280 个像素，在计算机中占用的大小为：1920 * 1280 * 3 *1 Bytes = 7 MB 的数据量

    - 在实际存储时，RGB 图像通常采用压缩格式来减小文件大小。常见的压缩格式包括 JPEG、PNG、BMP 等。这些格式可以将图像数据进行压缩，减少存储空间的占用，在计算机磁盘中看到的大小通常会小于原始图像的大小。

通道的意义

cv计算的接大部分是图像特征图中的通道的关系，图像特征图是由多个通道组成的，每个通道都代表了图像的一个特征。

特征图是神经网络中间层输出的结果，每个特征图都包含了多个通道，每个通道都代表了图像的一个特征。

RGB图像的通道数为3，每个像素点就有3个值，分别代表了图像的三个特征：红色、绿色和蓝色。

每个通道或多或少保留原始图像某些细节和轮廓特征，像上面3张分量图片一样，主要特征为颜色，R通道有着B通道没有的红色特征。
