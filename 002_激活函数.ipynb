我们把最左边的一列称为输入层，最右边的一列称为输出层，中间的一列称为中间层。中间层有时也称为隐藏层。​“隐藏”一词的意思是，隐藏层的神经元（和输入层、输出层不同）肉眼看不见.

一般而言，朴素感知机是指单层网络，即只有输入层和输出层，而没有隐层。指的是激活函数用了阶跃函数。
多层感知机指的神经网络，一般使用sigmoid函数等平滑的作为激活函数的多层网络


阶跃函数：当输入超过0时，输出1，否则输出0。
阶跃函数以0为界，输出从0切换为1（或者从1切换为0）​。它的值呈阶梯式变化，所以称为阶跃函数



```python
import numpy as np
import matplotlib.pyplot as plt
def step_function(x):
    return np.array(x>0,dtype=np.int32)  #数组中的数大于0为ture（1）,小于0为false（0），然后将将布尔值转换为整数
x=np.arange(-5.0,5.0,0.1)
y=step_function(x)
plt.plot(x,y)
plt.ylim(-0.1,1.1)    #指定y轴的范围
plt.show()
```

激活函数：

sigmoid 函数：
$$\sigma(z) = \frac{1}{1+e^{-z}}$$
$$\sigma'(z) = \sigma(z)(1-\sigma(z))$$

神经网络中用sigmoid函数作为激活函数，进行信号的转换，转换后的信号被传送给下一个神经元。

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

x=np.arange(-5,5,0.1)
y=sigmoid(x)
plt.plot(x,y)
plt.ylim(-0.1,1.1)
plt.show()
```

阶跃函数和sigmoid函数的不同点：

1. 平滑性不同：阶跃函数在0处不连续，sigmoid函数在0处连续。
2. 输出范围不同：阶跃函数的输出范围是0或1，sigmoid函数的输出范围是0到1。
3. 应用场景不同：阶跃函数通常用于分类问题，sigmoid函数通常用于回归问题。
4. 计算复杂度不同：阶跃函数的计算复杂度比sigmoid函数高。
5. 可导性不同：阶跃函数不可导，sigmoid函数可导。
6. 单调性不同：阶跃函数不单调，sigmoid函数单调。
7. 非线性性不同：阶跃函数不具有非线性性，sigmoid函数具有非线性性。  

共性：

1. 都是非线性函数。
2. 都是将输入映射到输出的函数。当输入信号为重要信息时，阶跃函数和sigmoid函数都会输出较大的值；当输入信号为不重要的信息时，两者都输出较小的值
3. 都是将输入映射到0到1之间的函数。不管输入信号有多小，或者有多大，输出信号的值都在0到1之间。

ReLU函数：
$$
f(x)=max(0,x)
$$




```python
def relu(x):
    return np.maximum(0, x)
x=np.arange(-1, 2, 1)  # 生成[-1,0,1]的数组
y=relu(x)
plt.plot(x,y)
plt.ylim(-1.1,2.1)    #指定y轴的范围
plt.show()
```

多维数组的运算

多维数组就是“数字的集合”，数字排成一列的集合、排成长方形的集合、排成三维状或者（更加一般化的）N维状的集合都称为多维数组。下面我们就用NumPy来生成多维数组，先从前面介绍过的一维数组开始。

```python
import numpy as np
a=np.array([1,2,3])
print(a)
print(a.shape) # 数组的形状 
print(a.size) # 数组的元素个数,这里的A.shape的结果是个元组(tuple)。这是因为一维数组的情况下也要返回和多维数组的情况下一致的结果
print(a.itemsize) # 数组中每个元素的字节大小
print(a.dtype)  # 数组元素的类型
print(a.ndim) # 数组的维数
```

二维数组：二维数组也称为矩阵(matrix)

```python
import numpy as np
b=np.array([[1,2,3],[4,5,6]])
print(b)
print(b.shape) # 表示第一个维度有2个元素，第二个维度有3个元素，另外，第一个维度对应第0维，第二个维度对应第1维（Python的索引从0开始）
print(b.size)  # 表示有6个元素
print(b.itemsize)
print(b.dtype)
print(b.ndim)
```

```python
#矩阵乘法
import numpy as np
a=np.array([[1,2,3],[3,4,5]])
b=np.array([[5,6],[7,8],[9,10]])
print(a)
print(b)
print(a.dot(b)) # 矩阵乘法
print(b.dot(a)) #调整a.b 顺序计算结果不一样
print (np.dot(a,b)) # 矩阵乘法
```
